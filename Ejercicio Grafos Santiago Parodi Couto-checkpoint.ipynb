# Visualización grafos con Networkx.

## Grafos con uniones de Pais, CCAA y Provincias

Se realizara un Grafo con las conecciones entre las distintas Comunidades Autonomas con España y las distintas provincias con sus respectivas comunidades.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
CCAA = nx.Graph()

CCAA.add_nodes_from(['España','Pais Vasco','Cast-La Mancha','Valenciana','Andalucia'\
                  ,'Castilla Leon','Extremadura','Baleares','Cataluña','Galicia'\
                  ,'Aragon','CCAA La Rioja','CCAA Madrid','Region Murcia','CCAA Navarra'\
                  ,'CCAA Asturias','Canarias','CCAA Cantabria','CCAA Ceuta','CCAA Melilla'])
```

```python
CCAA.add_edge('España', 'Pais Vasco')
CCAA.add_edge('España', 'Cast-La Mancha')
CCAA.add_edge('España', 'Valenciana')
CCAA.add_edge('España', 'Andalucia')
CCAA.add_edge('España', 'Castilla Leon')
CCAA.add_edge('España', 'Extremadura')
CCAA.add_edge('España', 'Baleares')
CCAA.add_edge('España', 'Cataluña')
CCAA.add_edge('España', 'Galicia')
CCAA.add_edge('España', 'Aragon')
CCAA.add_edge('España', 'CCAA La Rioja')
CCAA.add_edge('España', 'CCAA Madrid')
CCAA.add_edge('España', 'Region Murcia')
CCAA.add_edge('España', 'CCAA Navarra')
CCAA.add_edge('España', 'CCAA Asturias')
CCAA.add_edge('España', 'Canarias')
CCAA.add_edge('España', 'CCAA Cantabria')
CCAA.add_edge('España', 'CCAA Ceuta')
CCAA.add_edge('España', 'CCAA Melilla')


nx.draw(CCAA,
        with_labels=True,
        edge_color='lightblue',
        width=3,
        linewidths=100000,
        node_size=300,
        node_color='Yellow',
        alpha=0.8,
        )
```

```python
P = nx.Graph()
```

```python
P.add_nodes_from(['Ceuta', 'Melilla','Almeria','Cadiz','Cordoba'\
                  ,'Granada','Huelva','Jaen','Malaga','Sevilla'\
                  ,'Huesca','Teruel','Zaragoza','Las Palmas','Tenerife'\
                  ,'Cantabria','Canarias','Cantabria','Avila','Burgos'\
                  ,'Leon','Palencia','Salamanca','Segovia','Soria'\
                  ,'Valladolid','Zamora','Albacete','Ciudad Real','Cuenca'\
                  ,'Guadalajara','Toledo','Barcelona','Girona','Lleida'\
                  ,'Tarragona','Badajoz','Caceres','A Coruña','Lugo'\
                  ,'Ourense','Pontevedra','Islas Baleares','La Rioja','Alava'\
                  ,'Guipuzcoa','Vizcaya','Madrid','Navarra','Alicante'\
                  ,'Castellon','Valencia','Asturias','Murcia'])
```

```python
P.add_edge('Alava', 'Pais Vasco')
P.add_edge('Guipuzcoa', 'Pais Vasco')
P.add_edge('Vizcaya', 'Pais Vasco')
P.add_edge('Albacete', 'Cast-La Mancha')
P.add_edge('Ciudad Real', 'Cast-La Mancha')
P.add_edge('Cuenca', 'Cast-La Mancha')
P.add_edge('Guadalajara', 'Cast-La Mancha')
P.add_edge('Toledo', 'Cast-La Mancha')
P.add_edge('Alicante', 'Valenciana')
P.add_edge('Castellon', 'Valenciana')
P.add_edge('Valencia', 'Valenciana')
P.add_edge('Almeria', 'Andalucia')
P.add_edge('Cadiz', 'Andalucia')
P.add_edge('Cordoba', 'Andalucia')
P.add_edge('Granada', 'Andalucia')
P.add_edge('Huelva', 'Andalucia')
P.add_edge('Jaen', 'Andalucia')
P.add_edge('Malaga', 'Andalucia')
P.add_edge('Sevilla', 'Andalucia')
P.add_edge('Avila', 'Castilla Leon')
P.add_edge('Burgos', 'Castilla Leon')
P.add_edge('Leon', 'Castilla Leon')
P.add_edge('Palencia', 'Castilla Leon')
P.add_edge('Salamanca', 'Castilla Leon')
P.add_edge('Segovia', 'Castilla Leon')
P.add_edge('Soria', 'Castilla Leon')
P.add_edge('Valladolid', 'Castilla Leon')
P.add_edge('Zamora', 'Castilla Leon')
P.add_edge('Badajoz', 'Extremadura')
P.add_edge('Caceres', 'Extremadura')
P.add_edge('Islas Baleares', 'Baleares')
P.add_edge('Barcelona', 'Cataluña')
P.add_edge('Girona', 'Cataluña')
P.add_edge('Lleida', 'Cataluña')
P.add_edge('Tarragona', 'Cataluña')
P.add_edge('A Coruña', 'Galicia')
P.add_edge('Lugo', 'Galicia')
P.add_edge('Ourense', 'Galicia')
P.add_edge('Pontevedra', 'Galicia')
P.add_edge('Huesca', 'Aragon')
P.add_edge('Teruel', 'Aragon')
P.add_edge('Zaragoza', 'Aragon')
P.add_edge('La Rioja', 'CCAA La Rioja')
P.add_edge('Madrid', 'CCAA Madrid')
P.add_edge('Murcia', 'Region Murcia')
P.add_edge('Navarra', 'CCAA Navarra')
P.add_edge('Asturias', 'CCAA Asturias')
P.add_edge('Las Palmas', 'Canarias')
P.add_edge('Tenerife', 'Canarias')
P.add_edge('Cantabria', 'CCAA Cantabria')
P.add_edge('Ceuta', 'CCAA Ceuta')
P.add_edge('Melilla', 'CCAA Melilla')
```

```python
F = nx.compose(CCAA, P)
nx.draw(F,with_labels=True,
        edge_color='lightblue',
        width=3,
        linewidths=100000,
        node_size=300,
        node_color='Yellow',
        alpha=0.8,
        font_size=8)
```

```python
nx.draw_circular(F,with_labels=True,
        edge_color='Gray',
        width=3,
        linewidths=100000,
        node_size=300,
        node_color='lightblue',
        alpha=0.8,
        font_size=8)
```
