```python
inf_arg['Pais']='Arg'
inf_eur['Pais']='Eur'
inf_usa['Pais']='Usa'
inf_arg=inf_arg.reset_index()
inf_eur=inf_eur.reset_index()
inf_usa=inf_usa.reset_index()
```

```python
inf_arg2=inf_arg[(inf_arg['Date'] >= '1990-01-01') & (inf_arg['Date'] <= '2013-12-31')]
inf_eur2=inf_eur[(inf_eur['Date'] >= '1990-01-01') & (inf_eur['Date'] <= '2013-12-31')]
inf_usa2=inf_usa[(inf_usa['Date'] >= '1990-01-01') & (inf_usa['Date'] <= '2013-12-31')]
```

# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(inf_arg2['Date'], inf_arg2['Value'], label='Arg')
plt.plot(inf_eur2['Date'], inf_eur2['Value'], label='Eur')
plt.plot(inf_usa2['Date'], inf_usa2['Value'], label='Usa')
plt.xlabel('Fecha')
plt.ylabel('Inflacion')
plt.legend()
```

# Plotly

```python
inf_arg2["Categoria"] = pd.cut(inf_arg2.Value,bins=[2.42,43.53,84.63,125.73,166.9],labels=['A:>2.43','B:>43.53','C:>84.63','D:>125.73'])
inf_usa2["Categoria"] = pd.cut(inf_usa2.Value,bins=[127.39,154.08,180.77,207.46,234.2],labels=['A:>127.39','B:>154.08','C:>180.77','D:>207.46'])
inf_eur2["Categoria"] = pd.cut(inf_eur2.Value,bins=[69.45,77.11,84.77,92.43,100.2],labels=['A:>69.45','B:>77.11','C:>84.77','D:>92.43'])
```

```python
inf_arg3=inf_arg2.groupby('Categoria').count()
inf_usa3=inf_usa2.groupby('Categoria').count()
inf_eur3=inf_eur2.groupby('Categoria').count()
inf_arg3=inf_arg3.reset_index()
inf_eur3=inf_eur3.reset_index()
inf_usa3=inf_usa3.reset_index()
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
fig= px.histogram(inf_arg3, x= 'Categoria', y='Value').update_xaxes(categoryorder = 'trace')
fig.update_layout(title='Argentina')
fig.show()
```

```python
fig= px.histogram(inf_usa3, x= 'Categoria', y='Value').update_xaxes(categoryorder = 'trace')
fig.update_layout(title='USA')
fig.show()
```

```python
fig= px.histogram(inf_eur3, x= 'Categoria', y='Value').update_xaxes(categoryorder = 'trace')
fig.update_layout(title='Europa')
fig.show()
```

# Bokeh

```python
from bokeh.plotting import figure, output_file, save, gridplot, show
```

```python
output_file('bokeh_Santiago_Parodi_Couto.html')
s1 = figure(title="Grafico Linea Inflacion", x_axis_label='Fecha', y_axis_label='Inflacion')
s1.line(inf_arg2['Date'], inf_arg2['Value'], color='navy')

s2 = figure(x_range=s1.x_range, y_range=s1.y_range, title="Grafico Linea Inflacion", x_axis_label='Fecha', y_axis_label='Inflacion') #share both axis range
s2.line(inf_eur2['Date'], inf_eur2['Value'], color='red')

s3 = figure(x_range=s1.x_range, title="Grafico Linea Inflacion", x_axis_label='Fecha', y_axis_label='Inflacion') #share only one axis range
s3.line(inf_usa2['Date'], inf_usa2['Value'], color='green')

p = gridplot([[s1, s2, s3]])
show(p)
save(p)
```
