```python
import pandas as pd
```

```python
df1=pd.read_excel('Appol Datos.xlsx')
df1
```

```python
df2=pd.read_excel('Appol Datos.xlsx',sheet_name='Tabla Productos')
df2
```

```python
df1= df1.drop(['Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9'], axis=1)
```

```python
df1[['Trimestre','Año']]=df1.Periodo.str.split(' ',expand=True)
df1['Trim']=df1['Trimestre'].replace({"Q1": '15/02/', "Q2": '15/05/', "Q3": '15/08/', "Q4": '15/11/'})
df1['Fecha']=df1.Trim+df1.Año
df1=df1.drop(['Periodo','Trimestre','Trim','Año'], axis=1)
```

```python
df1[['Region2','Pais']]=df1.Region.str.split('/',expand=True)
df1[['Pais Abreviado','Continente']]=df1.Region2.str.split('-',expand=True)
df1=df1.drop(['Region','Region2'],axis=1)
```

```python
df1=df1.round({'Ingresos': 2, 'Gastos': 2})
df1
```

```python
df3=pd.merge(df1,df2,on='Tipo Producto')
df3
```

```python
df3['Beneficio']=df3['Ingresos']-df3['Gastos']
df3['Margen Beneficio']= df3['Beneficio']/df3['Ingresos'] *100
```

```python
df3 = df3[['Tipo Producto','Categoria Producto','Continente','Pais','Pais Abreviado','Fecha','Ingresos','Gastos','Beneficio','Margen Beneficio']]
```
